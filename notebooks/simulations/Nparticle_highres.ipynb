```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable
from matplotlib.lines import Line2D
import warnings; warnings.simplefilter('ignore')
import sys
import h5py
import pandas as pd
import seaborn as sns
sys.path.insert(0, '/Users/jsmonzon/Research/SatGen/mcmc/src/')
import jsm_ancillary
import jsm_visualize
import jsm_SHMR
import jsm_mcmc
import jsm_stats
import jsm_models
import evolve as ev
import galhalo as gh
import profiles as profiles
import config as cfg
from scipy.spatial import cKDTree
import pygtc
import jsm_simload
```

```python
plt.style.use('../../../SatGen/notebooks/paper1/paper.mplstyle')
double_textwidth = 7.0 #inches
single_textwidth = 3.5 #inches
levelz = [1-0.99, 1-0.95, 1-0.68]
```

```python
# npart = np.linspace(50, 1000, 10)
npart = np.array([int(50 / (6.2e6/1.5e8)), int(500 / (6.2e6/1.5e8)), int(1000 / (6.2e6/1.5e8))])

# npart = np.linspace(50, 1000, 10)
```

```python
tables = []

for n in npart:
    print(n)
    
    cat = jsm_simload.VSMDPL_HaloCatalogue(
        "../../../misc/ForSebastian/full/N_AB_sub_vsmdpl.dat",
        npart_thresh=n,
        xoff_thresh=0.07,
        spin_thresh=0.07,
        isolation_factor=3)
    
    tables.append(cat._build_host_table(sample="isolated"))
```

```python
rho_matz, rho_err_matz, N_hosts = jsm_simload.mass_binned_correlation(tables, xkey="log1pz50", ykey="logNsub", xlabel="log(1+z50)", ylabel="logN", make_plot=False)

rho_matc, rho_err_matc, _ = jsm_simload.mass_binned_correlation(tables, xkey="logc", ykey="logNsub", xlabel="log c", ylabel="logN", make_plot=False)

rho_matm, rho_err_matm, _ = jsm_simload.mass_binned_correlation(tables, xkey="logMMs", ykey="logNsub", xlabel="log MMs", ylabel="logN", make_plot=True)
```

```python
logMvir_binned = np.linspace(12.6, 14, 8)

cmap = plt.cm.viridis
norm = Normalize(vmin=npart.min(), vmax=npart.max())

fig, ax = plt.subplots(3, 1, figsize=(double_textwidth, 1.5*single_textwidth), sharex=True, sharey=True)

for k, rho_arr in enumerate(rho_matz):
    ax[0].plot(logMvir_binned, np.abs(rho_arr), marker=".", color=cmap(norm(npart[k])))

for k, rho_arr in enumerate(rho_matc):
    ax[1].plot(logMvir_binned, np.abs(rho_arr), marker=".", color=cmap(norm(npart[k])))

for k, rho_arr in enumerate(rho_matm):
    ax[2].plot(logMvir_binned, np.abs(rho_arr), marker=".", color=cmap(norm(npart[k])))

ax[0].set_ylabel("$\\rho_z$")
ax[1].set_ylabel("$\\rho_c$")
ax[2].set_ylabel("$\\rho_m$")

ax[2].set_ylim(0, 0.8)
ax[2].set_xlim(12.6, 14.0)
ax[2].set_xlabel("$\\log \\rm M_{vir}$")

# add horizontal colorbar above top axis
sm = ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

# get position of top axis and place colorbar above it
pos = ax[0].get_position()
cbar_ax = fig.add_axes([pos.x0, pos.y1 + 0.02, pos.width, 0.03])
cbar = fig.colorbar(sm, cax=cbar_ax, orientation='horizontal')
cbar.set_label("$N_{\\rm particle}$")
cbar_ax.xaxis.set_ticks_position('top')
cbar_ax.xaxis.set_label_position('top')


plt.show()
```
