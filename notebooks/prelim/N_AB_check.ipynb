```python
N_AB_sub = np.loadtxt('../../../misc/ForSebastian/BolshoiBP/N_AB_sub_BolPlanck.dat')

cols = [
    "host_id", "logMh", "ch", "a_50h",
    "id", "log10Mvir", "Rvir", "rs", "vrms", "scale_of_last_MM",
    "vmax", "x", "y", "z", "vx", "vy", "vz",
    "Jx", "Jy", "Jz", "Spin", "Tidal_Force", "Tidal_ID",
    "Mmvir_all", "M200b", "M200c", "M500c", "Xoff", "Voff",
    "Spin_Bullock", "b_to_a", "c_to_a",
    "Ax", "Ay", "Az", "T_by_U",
    "M_pe_Behroozi", "M_pe_Diemer",
    "Macc", "Mpeak", "Vacc", "Vpeak", "Halfmass_Scale",
    "Acc_Rate_Inst", "Acc_Rate_100Myr", "Acc_Rate_1Tdyn",
    "Acc_Rate_2Tdyn", "Acc_Rate_Mpeak",
    "Mpeak_Scale", "Acc_Scale", "First_Acc_Scale",
    "First_Acc_Mvir", "First_Acc_Vmax", "Vmax_at_Mpeak",
    "Tidal_Force_Tdyn"
]

df = pd.DataFrame(N_AB_sub, columns=cols)

import numpy as np
import pandas as pd

def build_host_table(df, mass_thresh):

    host_id_unique = df["host_id"].unique()

    logMh  = np.zeros(len(host_id_unique))
    c_h    = np.zeros(len(host_id_unique))
    a_half = np.zeros(len(host_id_unique))
    Nsub   = np.zeros(len(host_id_unique))
    f_sub  = np.zeros(len(host_id_unique))

    for i, hid in enumerate(host_id_unique):

        subset  = df[df["host_id"] == hid]
        subset1 = subset[subset["log10Mvir"] >= mass_thresh]

        if len(subset) > 0:
            logMh[i]  = subset["logMh"].mean()
            c_h[i]    = subset["ch"].mean()
            a_half[i] = subset["a_50h"].mean()

            Nsub[i] = len(subset1)

            # convert log mass to linear mass
            host_mass = 10**logMh[i]
            sub_mass_total = np.sum(10**subset1["log10Mvir"])

            f_sub[i] = sub_mass_total / host_mass

        if (i % 1000) == 0:
            print(i)

    small_df = pd.DataFrame({
        "host_id": host_id_unique,
        "logMh": logMh,
        "c_h": c_h,
        "a_half": a_half,
        "Nsub": Nsub,
        "f_sub": f_sub
    })

    return small_df
```

```python
small_df = build_host_table(df, mass_thresh=9.82)
```

```python
small_df.to_csv("../../data/summary_tabs/bolshoi_full.csv")
```
